```python
import matplotlib.pyplot as plt

# understand the risk column
print("Risk column stats:")
print(df['risk'].describe())
print("\nNull values in risk:", df['risk'].isnull().sum())

# visualize distribution
df['risk'].hist(bins=50, figsize=(10,4))
plt.title('Distribution of Risk Values')
plt.xlabel('Risk (log scale)')
plt.ylabel('Count')
plt.show()
```

```python
import numpy as np

def classify_risk(risk):
    if risk > -5:
        return 'HIGH'
    elif risk > -10:
        return 'MEDIUM'
    else:
        return 'LOW'

df['risk_label'] = df['risk'].apply(classify_risk)
print(df['risk_label'].value_counts())
print("\nPercentage:")
print(df['risk_label'].value_counts(normalize=True) * 100)
```

```python
# Check null values
null_counts = df.isnull().sum()
null_counts = null_counts[null_counts > 0].sort_values(ascending=False)

print(f"Total columns with nulls: {len(null_counts)}")
print("\nTop 20 columns with most nulls:")
print(null_counts.head(20))
```

```python
# Drop high null column and identity columns
df = df.drop(columns=['c_rcs_estimate', 'event_id', 'mission_id'])

# Fill remaining nulls with median
df = df.fillna(df.median(numeric_only=True))

# Verify no nulls remain
print("Remaining nulls:", df.isnull().sum().sum())
print("Dataset shape after cleaning:", df.shape)
```

```python
# Define our features based on domain knowledge
features = [
    # Core conjunction metrics
    'time_to_tca',
    'miss_distance',
    'relative_speed',
    'mahalanobis_distance',

    # Relative position components
    'relative_position_r',
    'relative_position_t',
    'relative_position_n',

    # Relative velocity components
    'relative_velocity_r',
    'relative_velocity_t',
    'relative_velocity_n',

    # Target object uncertainty
    't_sigma_r', 't_sigma_t', 't_sigma_n',
    't_sigma_rdot', 't_sigma_tdot', 't_sigma_ndot',

    # Chaser object uncertainty
    'c_sigma_r', 'c_sigma_t', 'c_sigma_n',
    'c_sigma_rdot', 'c_sigma_tdot', 'c_sigma_ndot',

    # Orbital parameters
    't_j2k_sma', 't_j2k_ecc', 't_j2k_inc',
    'c_j2k_sma', 'c_j2k_ecc', 'c_j2k_inc',

    # Observation quality
    't_obs_used', 'c_obs_used',
    't_weighted_rms', 'c_weighted_rms',

    # Altitude
    't_h_apo', 't_h_per',
    'c_h_apo', 'c_h_per',

    # Space weather
    'F10', 'AP'
]

# Create feature matrix X and target vector y
X = df[features]
y = df['risk_label']

print("Feature matrix shape:", X.shape)
print("Target distribution:")
print(y.value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 80% train, 20% test
    random_state=42,    # reproducibility
    stratify=y          # maintain class proportions in both splits
)

print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
print("\nTraining class distribution:")
print(y_train.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
import time

print("Training started...")
start = time.time()

model = RandomForestClassifier(
    n_estimators=100,        # 100 decision trees
    class_weight='balanced', # handle our imbalance problem
    random_state=42,         # reproducibility
    n_jobs=-1                # use all CPU cores, trains faster
)

model.fit(X_train, y_train)

end = time.time()
print(f"Training complete in {round(end-start, 2)} seconds")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Make predictions on test set
y_pred = model.predict(X_test)

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=['LOW', 'MEDIUM', 'HIGH'])

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=['LOW', 'MEDIUM', 'HIGH'],
            yticklabels=['LOW', 'MEDIUM', 'HIGH'],
            cmap='Blues')
plt.title('VoidWatch — Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
# Feature importance
import pandas as pd
import matplotlib.pyplot as plt

importances = pd.Series(
    model.feature_importances_,
    index=features
).sort_values(ascending=False)

plt.figure(figsize=(10,8))
importances.head(15).plot(kind='barh')
plt.title('VoidWatch — Top 15 Most Important Features')
plt.xlabel('Importance Score')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

print("\nTop 10 features:")
print(importances.head(10))
```
